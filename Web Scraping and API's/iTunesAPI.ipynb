```python
import requests
import json
import pandas as pd
```

```python
base_site = "https://itunes.apple.com/search"
```

```python
url = base_site + "?term=the+beatles&country=us"
```

```python
requests.get(url)
```

```python
r = requests.get(base_site, params = {"term":"the beatles", "country":"us", "limit":"200"})
```

```python
r.status_code
```

```python
info = r.json()
print(json.dumps(info, indent = 4))

```

```python
info.keys()
```

```python
(info['results'][1]["artistName"])
```

```python
info['results'][1].keys()
```

```python
info['results'][1]['trackName']
```

```python
for i in range(len(info['results'])):
    if 'artistName' in info['results'][i].keys():
        print("The artist:" + " " + info['results'][i]['artistName'])
        print("The release data: "+ info['results'][i]['releaseDate'])
```

```python
songs_df = pd.DataFrame(info['results'])
```
