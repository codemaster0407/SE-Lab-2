Using movie reviews dataset from nltk.corpus as my testing data

Storing all the words in each reviews to documents

```python
documents = []


for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids():
        documents.append((movie_reviews.words(fileid)))
```

```python
type(documents)
```

```python
all_words = []

for doc in documents:
    for words in doc:
        all_words.append(words)
```
