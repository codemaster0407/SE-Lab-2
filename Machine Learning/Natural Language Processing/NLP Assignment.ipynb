```python
import string
from string import punctuation
```

```python
punctuations = string.punctuation
```

```python
punctuations
```

Now we write a function to remove all the punctuation marks from the total words list

```python
def remove_punctuations(words, punctuations):
    for i in words:
        if i in punctuations:
            words.remove(i)
            
    return words
            
            
```

```python
words = ["Hello","#", "Natural", "NLP"]
```

```python
remove_punctuations(words,punctuations)
```

Just an example of how it removes the punctuations from the words 
Now applying it for all the words

```python
words_webtext = remove_punctuations(words_webtext, punctuations)
```

```python
def getUniqueWords(allWords) :
    uniqueWords = [] 
    for i in allWords:
        if not i in uniqueWords:
            uniqueWords.append(i)
    return uniqueWords
```

```python
uniquewords = getUniqueWords(words_webtext)
```

```python
len(uniquewords)
```

Checking Heap's Law for the above vocabulary.

V(n) = k * n^β  
log V(n) = β * log(n) + k


Checking with different size of words from web_text

```python
#Taking 3000 and 6000 as different sizes for words
lentotalwords1 = 3000
lentotalwords2 = 6000

temp_words1 = []
for i in range(lentotalwords1):
    n = random.randint(0,len(words_webtext))
    temp_words1.append(words_webtext[n])
    
    
```

```python
temp_words2 = []

for i in range(lentotalwords2):
    n = random.randint(0,len(words_webtext))
    temp_words2.append(words_webtext[n])
```

Finding the number of unique words in both temp_words1 and temp_words2

```python

print("Total number of words ->",len(temp_words1), len(temp_words2))

temp_words1 = getUniqueWords(temp_words1)
temp_words2 = getUniqueWords(temp_words2)

lenunique1 = len(temp_words1)
lenunique2 = len(temp_words2)

print("Unique words ->" ,len(temp_words1), len(temp_words2))
```

Using Heap's Law

```python
beta = (math.log(lenunique2) - math.log(lenunique1)) / (math.log(lentotalwords2) - math.log(lentotalwords1))
beta
```

```python
# K1 = math.log(lenunique2) - ( beta * math.log(lentotalwords2))
K = math.log(lenunique1) - ( beta * math.log(lentotalwords1))
print(K)
```

```python
unique_words_heap = K* (len(words_webtext)**beta)

unique_words_heap
```
