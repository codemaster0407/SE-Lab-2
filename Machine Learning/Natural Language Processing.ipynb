## Stemming

```python
stem_words = ["play","playing","player","played"]
ps = PorterStemmer()

stemmed_words = [ps.stem(w) for w in stem_words]

stemmed_words
```

```python
from nltk.corpus import wordnet

def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```
